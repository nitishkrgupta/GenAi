```python
from nltk.corpus import stopwords

stopwords = stopwords.words('english')
```

```python
with open('D:\Study\pyspider\GenAI\Text_normalization\data.txt', 'r') as file:
    data = file.read()

print(data)
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
tokens = word_tokenize(data)
tokens
```

```python
updated_tokens = []
for word in tokens:
    if word not in stopwords:
        updated_tokens.append(word)
print(updated_tokens)
```

```python
nlp_sentence = [
    "Python is a popular programming language for NLP.",
    "NLP enables computers to understand and generate human language.",
    "Tokenization splits text into individual words or sentences.",
    "Stopword removal often improves text analysis by removing common words.",
    "Stemming and lemmatization reduce words to their base forms.",
    "Part-of-speech tagging assigns grammatical roles to each word.",
    "Named entity recognition finds people, organizations, and locations in text.",
    "Word embeddings map words to dense vector representations.",
    "Text classification assigns categories or labels to text documents.",
    "Pretrained models like BERT and GPT have advanced many NLP tasks."
]
```

```python
updated_sent = []

for sent in nlp_sentence:
    s1 = ""
    for word in sent.split():
        if word not in stopwords:
            s1 += word + " "
    updated_sent.append(s1.strip())  
print(updated_sent)
```

```python
for index in range(len(nlp_sentence)):
    words = nlp_sentence[index].split()
    # print(words)
    filtered_words = [word for word in words if word not in stopwords]
    nlp_sentence[index] = " ".join(filtered_words)
print(nlp_sentence)
```
