```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
words = [
    "Playing", "Played", "player", "running", "runner",
    "Studies", "Studying", "caring", "cared", "happiness",
    "easily", "flying", "connected", "connection", "connections",
    "reading", "reader", "writing", "written", "organization", 'a', 'the'
]

for word in words:
    print(f"{word} ---> {stemmer.stem(word)}")
```

```python
from nltk.stem import SnowballStemmer

snowball_stemmer = SnowballStemmer(language='english', ignore_stopwords=True)
```

```python
snowball_words = [
    "Playing", "Played", "player", "running", "runner",
    "Studies", "Studying", "caring", "cared", "happiness",
    "easily", "flying", "connected", "connection", "connections",
    "reading", "reader", "writing", "written", "organization", 'a', 'the'
]
```

```python
for word in snowball_words:
    print(f"{word} ---> {snowball_stemmer.stem(word)}")
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer


stop_words = stopwords.words("english")
print(stop_words)
snowball_stemmer = SnowballStemmer('english', ignore_stopwords=True)
snowball_stemmer.stopwords = stop_words
# snowball_stemmer.stopwords
for word in snowball_words:
    print(word, snowball_stemmer.stem(word))
```
