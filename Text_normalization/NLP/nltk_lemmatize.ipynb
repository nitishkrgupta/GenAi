```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
# nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
tokens = ["running", "studies","children", "better", "went", "mice", "cars", "eating", "leaves","written"]
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for word in tokens:
    print(word, lemmatizer.lemmatize(word, pos='n'))
# n for noun
# nltk lemmatization has limitation it only change the one part of speech at a time
```

```python
'''
The Part Of Speech tag. Valid options are "n" for nouns, "v" for verbs, "a" for adjectives, "r" for adverbs and "s" for satellite adjectives
'''
```

```python
nltk.pos_tag(tokens) # to check part of speech of tokens

'''
N - 'noun'
V - 'verb'
J - 'a'- 'adjective'
R - '
'''
```

```python
def find_postag(tag):
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('N'):
        return 'n'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('R'):
        return 'r'
    else:
        return 'n'
```

```python
pos_tag_word = nltk.pos_tag(tokens)
pos_tag_word
```

```python
for word, tag in pos_tag_word:
    tagged_word = lemmatizer.lemmatize(word, find_postag(tag))
    print(tagged_word)
```
