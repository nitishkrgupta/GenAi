```python
import spacy

nlp = spacy.load('en_core_web_sm')

tokens = nlp('''Apply the pipeline to some text. The text can span multiple sentences, and can contain arbitrary whitespace. 
             Alignment into the original string is preserved. ''')
print(type(tokens), tokens)
# by default it converts into tokens it doesn't required tokenization
```

```python
lemmatize_tokens = [token.lemma_ for token in tokens if not token.is_stop]
# for token in tokens:
#     print(token.lemma_, token.is_stop)
lemmatize_tokens
print(" ".join(lemmatize_tokens))
```

```python
text = '''Narendra Modi flies On 21-07-2026, Bengaluru to Bangkok to get Thai massage from Professionals, flight price is 5000. 2004, Instagram, ticket sent to kumarknitish@gmail.com'''
tokens = nlp(text)
```

```python
tokens.ents
```

```python
for ent in tokens.ents:
    print(ent.text, ent.label_)
```

```python
for token in tokens:
    print(token.pos_, token.tag_)
```
