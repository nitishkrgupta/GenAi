```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
res = '''            Return a tokenized copy of text, using NLTK's recommended word tokenizer (currently an improved ." \
                          TreebankWordTokenizer along with .PunktSentenceTokenizer for the specified language).'''
```

```python
word_tokens = word_tokenize(res)
word_tokens
```

```python
sent_tokens = sent_tokenize(res)
sent_tokens
```
