```python
import string

print(string.punctuation)
print(len(string.punctuation))
```

```python
# clean the text by removing punctuation
with open('data.txt', 'r') as fh:
    data = fh.read()    
    data_new = ""
    for i in data:
        if i not in string.punctuation:
            data_new += i
with open('data.txt', 'w') as fh:
    fh.write(data_new)
```
