#### fixed size chunking

```python
with open('file.txt', 'r') as file:
    data = file.read()
data
```

```python
fixed_size = 20
chunks = []
for i in range(0, len(data), fixed_size):
    chunks.append(data[i:i+fixed_size])
chunks
```

#### Fixed size + overlapping

```python
fixed_size = 20
overlapping = 4
chunks = []
for i in range(0, len(data), fixed_size-overlapping):
    chunks.append((data[i:i+fixed_size]))
chunks
```

#### Paragraph based chunking

```python
import re

chunks = re.split('\n\n', data)

chunks
```
